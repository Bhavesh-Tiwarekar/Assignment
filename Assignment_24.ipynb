##### Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
##### Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
## Program to create a line plot using the 'fmri' dataset.

import seaborn as sns

import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")

# Plot the line plot using the "timepoint" as x-axis and "signal" as y-axis
sns.lineplot(x= "timepoint", y = "signal", data= fmri, hue = "event", style= "region")
plt.title("FMRI Data")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title = "Event")
plt.show()
```

##### Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
##### Note: pclass, age, and fare are columns in the titanic dataset.

```python
## Program to plot two box plots using x = 'pclass', y = 'age' and y = 'fare' from the "titanic" dataset.

titanic = sns.load_dataset("titanic")

sns.set(style="whitegrid")

# Create the first box plot for age distribution by passenger class
plt.subplot(121)
sns.boxplot(data= titanic, x = "pclass", y = "age")
plt.title("Age Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")

# Create the second box plot for fare distribution by passenger class
plt.subplot(122)
sns.boxplot(data= titanic, x = "pclass", y = "fare")
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.tight_layout()

# Show the plots
plt.show()
```

##### Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
## Program to plot a histogram using the "diamonds" dataset.

diamonds = sns.load_dataset("diamonds")

sns.histplot(x = "price", data= diamonds, hue= "cut", multiple= "stack")

plt.title("Price Distribution of Diamonds by Cut")
plt.xlabel("Price")
plt.ylabel("Frequency")

# Show the plot
plt.show()
```

##### Q5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
## Program to create a pair plot using the "iris" dataset.

iris = sns.load_dataset("iris")

sns.set_style(style="ticks")
sns.pairplot(iris, hue= "species")

# Show the plot
plt.show()
```

##### Q6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
## Program to create a heatmap from the "flights" dataset.

flights = sns.load_dataset("flights")

pivot_flights = flights.pivot_table(index= "year", columns="month", values= "passengers")

# Create the heatmap using the pivoted data
sns.heatmap(pivot_flights, annot=True, fmt='d', linewidths=0.5, annot_kws={"size": 8})
plt.title("Number of Passengers by Month and Year")
plt.xlabel("Month")
plt.ylabel("Year")

# Show the plot
plt.show()
```
