##### Q1. List any five functions of the pandas library with execution.

##### Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the
##### DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
## Program to re-index a DataFrame using a python function.

import pandas as pd

def new_index(df):
    new_index = range(1, (2 * len(df)) + 1, 2)
    df_copy = df.copy()
    df_copy.index = new_index
    return df_copy
    
data = {'A': [10, 20, 30, 40],
        'B': [50, 60, 70, 80],
        'C': [90, 100, 110, 120]}

df = pd.DataFrame(data)

new_index(df)
```

##### Q3. You have a Pandas DataFrame df with a column named 'Values'. Write a Python function that iterates over the DataFrame and calculates the sum of the 
##### first three values in the 'Values' column. The function should print the sum to the console.
##### For example, if the 'Values' column of df contains the values [10, 20, 30, 40, 50], your function should
##### calculate and print the sum of the first three values, which is 60.

```python
## Program to create a function to calculate the sum of first three values of Values column in a DataFrame 

import pandas as pd

def df_sum_first_three_values(df):
    count = 0
    sum = 0
    for i in df["Values"]:
        sum += i
        count += 1
        if count == 3:
            break
    return sum

data = {'Values': [10, 20, 30, 40, 50]}

df = pd.DataFrame(data)

print(f"The sum of first three values of DataFrame is {df_sum_first_three_values(df)}")
```

##### Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column
##### 'Word_Count' that contains the number of words in each row of the 'Text' column.

```python
## Program to create a function that can create a new column 'Word_Count' that contains the number of words in each row of the 'Text' column of a DataFrame.

import pandas as pd

def df_word_count(df):
    l = []
    for i in df["Text"]:
        l.append((len(i.split())))
    df["Word_Count"] = l
    return df

data = {'Text': ["This is a sample sentence.",
                 "Hello, how are you?",
                 "Python programming is fun!"]}

df = pd.DataFrame(data)

df_word_count(df)
```

##### Q7. You have a Pandas DataFrame df that contains a column named 'Email' that contains email addresses in the format 'username@domain.com'. Write a 
##### Python function that creates a new column 'Username' in df that contains only the username part of each email address.
##### The username is the part of the email address that appears before the '@' symbol. For example, if the email address is 'john.doe@example.com',
##### the 'Username' column should contain 'john.doe'. 
##### Your function should extract the username from each email address and store it in the new 'Username' column.

```python
## Program to create a function that extracts the username from each email address and store it in the new 'Username' column inside a DataFrame.

import pandas as pd

def extract_username(df):
    l1 = []
    for i in range(len(df)):
        l1.append(df["Email"][i][0:df["Email"][i].find("@")])
    df["Username"] = l1
    return df

email_addresses = {
    "Email": ["akshay@gmail.com", "anant@yahoo.com", "john123@rediff.com"]
}

df = pd.DataFrame(email_addresses)
  
extract_username(df)
```

##### Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects all rows where the value in 
##### column 'A' is greater than 5 and the value in column 'B' is less than 10. The function should return a new DataFrame that contains only the selected rows.
##### For example, if df contains the following values:
#####   A B C
##### 0 3 5 1
##### 1 8 2 7
##### 2 6 9 4
##### 3 2 3 5
##### 4 9 1 2

##### Your function should select the following rows: A B C
##### 1 8 2 7
##### 4 9 1 2
##### The function should return a new DataFrame that contains only the selected rows.

```python
## Program to create a function that return a new DataFrame that contains only the selected rows based on a condition.

import pandas as pd

def selected_rows(df):
    selected_rows = df[(df["A"] > 5) & (df["B"] < 10)]
    return selected_rows

data = {
    "A": [3,8,6,2,9],
    "B": [5,2,9,3,1],
    "C": [1,7,4,5,2]
}

df = pd.DataFrame(data)

new_selected_rows_df = selected_rows(df)

print(new_selected_rows_df)
```

##### Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean, median, and standard deviation of the 
##### values in the 'Values' column.

```python
## Program to create a function to calculate the mean, median, and standard deviation of the values in the 'Values' column of a DataFrame.

import pandas as pd

def df_statistics(df):
    df_mean = df["Values"].mean()
    df_median = df["Values"].median()
    df_std = df["Values"].std()
    
    return df_mean, df_median, df_std

data = {'Values': [10, 20, 30, 40, 50]}

df = pd.DataFrame(data)

Mean, Median, Std_deviation = df_statistics(df)

print("Mean:", Mean)
print("Median:", Median)
print("Standard Deviation:", Std_deviation)
```

##### Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' 
##### that contains the moving average of the sales for the past 7 days for each row in the DataFrame. 
##### The moving average should be calculated using a window of size 7 and should include the current day.

```python
## Program to create a function to calculate moving average of the sales for the past 7 days for each row in the DataFrame.

import pandas as pd

def calculate_moving_average(df):
    l = []
    for i in range(len(df["Sales"])):
        if i < 7:
            l.append(df["Sales"][0: i+1].mean())
        else: 
            l.append(df["Sales"][i: i-7: -1].mean())
    df["'MovingAverage"] = l
    return df

data = {
    'Date': pd.date_range(start='2023-08-01', periods=14),
    'Sales': [10, 20, 15, 30, 25, 40, 35, 50, 45, 60, 55, 70, 65, 80]
}

df = pd.DataFrame(data)

calculate_moving_average(df)
```

##### Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new column 'Weekday' in the DataFrame. 
##### The 'Weekday' column should contain the weekday name (e.g. Monday, Tuesday) corresponding to each date in the 'Date' column.
##### For example, if df contains the following values:
#####    Date
##### 0 2023-01-01
##### 1 2023-01-02
##### 2 2023-01-03
##### 3 2023-01-04
##### 4 2023-01-05
##### Your function should create the following DataFrame:

#####      Date    Weekday
##### 0 2023-01-01 Sunday
##### 1 2023-01-02 Monday
##### 2 2023-01-03 Tuesday
##### 3 2023-01-04 Wednesday
##### 4 2023-01-05 Thursday
##### The function should return the modified DataFrame.

```python
## Program to create a function that creates a new column 'Weekday' with weekday name corresponding to each row in 'Date' column in the DataFrame.

import pandas as pd

def df_weekday_name(df):
    l = []
    for i in range(len(df["Date"])):
        l.append(pd.Timestamp(df["Date"][i]).day_name())
    df["Weekday"] = l
    return df

data = {
    "Date": ["2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04", "2023-01-05"]
}

df = pd.DataFrame(data)

df_weekday_name(df)
```

##### Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is 
##### between '2023-01-01' and '2023-01-31'.

```python
## Program to create a function to select all rows where the date is between '2023-01-01' and '2023-01-31' from a DataFrame.

import pandas as pd

def df_select_rows_between_dates(df, start_date, end_date):
    selected_rows = df[(df["Date"] >= start_date) & (df["Date"] <= end_date)]
    return selected_rows

data = {
    'Date': pd.date_range(start='2023-01-01', end='2023-02-28', freq='D'),
}

df = pd.DataFrame(data)

start_date = "2023-01-01"
end_date = "2023-01-31"

df_select_rows_between_dates(df, start_date, end_date)
```
