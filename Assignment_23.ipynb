##### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
##### import numpy as np
##### np.random.seed(3)
##### x = 3 + np.random.normal(0, 2, 50)
##### y = 3 + np.random.normal(0, 2, len(x))
##### Note: Also add title, xlabel, and ylabel to the plot.

```python
"""
Answer:

1. A scatter plot uses dots to display values of two different numeric values.
2. It shows relationship between two variables.
3. Code to generate scatter plot is as follows:
"""
## Program to generate a scatter plot

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

import matplotlib.pyplot as plt

plt.scatter(x, y)
plt.xlabel("x data")
plt.ylabel("y data")
plt.title("Random data from a normal distribution")
plt.show()
```

##### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
##### Use the following data:
##### import numpy as np
##### For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
##### For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
##### For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
##### For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
"""
Answer:

1. The subplot() function adds subplot to a current figure at the specified grid position.
2. It allows to visualize multiple plots side by side or in a grid layout, making it easier to compare different visualizations or datasets.
3. Code to draw four line plots using subplot() function is as follows: 
"""

import numpy as np

# Plotting the lines using a 2 x 2 grid of subplots

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(221)
plt.plot(x1, y1)
plt.xlabel("x1")
plt.ylabel("y1")
plt.title('Line 1')

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(222)
plt.plot(x2, y2)
plt.xlabel("x2")
plt.ylabel("y3")
plt.title('Line 2')

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(223)
plt.plot(x3, y3)
plt.xlabel("x3")
plt.ylabel("y3")
plt.title('Line 3')

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(224)
plt.plot(x4, y4)
plt.xlabel("x4")
plt.ylabel("y4")
plt.title('Line 4')

# Adjust layout to avoid overlap
plt.tight_layout()

# Display the plots
plt.show()
```

##### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
##### import numpy as np
##### company = np.array(["Apple", "Microsoft", "Google", "AMD"])
##### profit = np.array([3000, 8000, 1000, 10000])

```python
"""
Answer:

1. Bar plots represent categorical data using rectangular bars.
2. They're useful for comparing different categories or groups.
3. Code to plot a bar plot and a horizontal bar plot with the given company and profit data:
"""

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


# Vertical bar graph
plt.bar(company, profit, color = 'blue')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Company Profits')
plt.show()

# horizontal bar graph
plt.barh(company, profit, color = 'green')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Company Profits')
plt.show()
```

##### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
##### box1 = np.random.normal(100, 10, 200)
##### box2 = np.random.normal(90, 20, 200)

```python
"""
Answer:

1. A box plot is a graphical representaion displaying variation in a set of data.
2. A box plot is particularly useful for visualizing the spread and variability of data, identifying potential outliers, and comparing multiple datasets.
3. It displays key statistical measures such as the minimum, first quartile, median, third quartile, and maximum.
4. Code to plot a box plot using the given dataset:
"""
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
## Program to plot a box plot using the given dataset.

plt.boxplot([box1, box2], labels=("Box1", "Box2"))
plt.xlabel('Boxes')
plt.ylabel('Values')
plt.title('Box Plot of Two Arrays')
plt.show()
```
