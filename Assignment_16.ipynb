```python
# Program to create an API and use Postman to test it.

from flask import Flask, jsonify

app = Flask(__name__)

@app.route("/data", methods = ['GET'])
def get_data():
    data = {
        'name': 'BT',
        'age': 25,
        'occupation': 'Developer'
    }
    
    return jsonify(data)

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
