#### Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

```python
"""
Answer:

1. In Pandas, a DataFrame is a 2-dimensional labeled data structure with columns that can be of different data types.
2. The difference between a Series and a DataFrame is that a DataFrame can hold multiple columns, while a Series can hold only one column of data.
"""

## Program to show difference between pandas dataframe and series.

import pandas as pd

# Using DataFrame
data = {
    "Name": ["Aditya", "John", "Shubham"],
    "Age": [25, 30, 22],
    "Grade": [85, 90, 78]
}

df = pd.DataFrame(data)
print("Dataframe:")
print(df)

# Using Series
names = pd.Series(["Aditya", "Bob", "Charlie"], name='Name')
ages = pd.Series([25, 30, 22], name='Age')
grades = pd.Series([85, 90, 78], name='Grade')

print("\nSeries - Names:")
print(names)
print("\nSeries - Ages:")
print(ages)
print("\nSeries - Grades:")
print(grades)

"""
In the above code, the 'df' DataFrame contains three columns: "Name", "Age", and "Grade". Each column is represented by a Series.
The 'names', 'ages', and 'grades' Series each represent a single column of data.
"""
```

#### Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can
#### you give an example of when you might use one of these functions?

```python
"""
Answer:

The following are some of the common functions used to manipulate data in Pandas DataFrame:
1. head() and tail(): These functions allow to view the first or last 5 rows of a DataFrame.
   Example:
"""
data = {
    "Name": ["Aditya", "John", "Shubham"],
    "Age": [25, 30, 22],
    "Grade": [85, 90, 78]
}

df = pd.DataFrame(data)

print(df.head(5))  # Show the first 5 rows
print(df.tail(3))  # Show the last 3 rows
"""
2. describe(): This function provides summary statistics of numeric columns in the DataFrame.
   Example:
"""
print("\n", df.describe())

"""
3. loc[] and iloc[]: These indexers allow you to access specific rows and columns by label (using row and column names) or by integer-based indexing respectively.
   Example:
"""

print("\n", df.loc[1:2, ["Name"]])   # Access a specific element by label and column name

print("\n", df.iloc[1, 2])           # Access a specific element by integer-based indexing

"""
4. sort_values(): This function sorts the DataFrame based on one or more columns.
"""
df.sort_values(by = "Age")
```

#### Q7. Create a DataFrame using multiple Series. Explain with an example.

```python
## Program to create a DataFrame using multiple Series

import pandas as pd

# Creating Series
names = pd.Series(["Aditya", "Bob", "Charlie"])
ages = pd.Series([25, 30, 22])
grades = pd.Series([85, 90, 78])

# Creating DataFrame using Series
data = {
    'Name': names,
    'Age': ages,
    'Grades': grades
}

df = pd.DataFrame(data)

print(df)

"""
In the above code, three series named 'names', 'ages', 'grades' were created.
A dictionary 'data' was created where the keys are column names and the values are the corresponding Series.
The 'data' dictionary was used to create the DataFrame 'df'.
"""
```
