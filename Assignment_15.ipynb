#### Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

```python
# Program to create a Flask application to display 'Hello World!!'.

from flask import Flask

app = Flask(__name__)

@app.route("/Q2_hello_world")
def hello_world():
    return "<h3>Hello World!!</h3>"

if __name__ == "__main__":
    app.run(host="0.0.0.0")
    
```

#### Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
##### Company Name: ABC Corporation
##### Location: India
##### Contact Detail: 999-999-9999
#### Attach the screenshot of the output in Jupyter Notebook.

```python
# Program to display welcome message and shoe company details

from flask import Flask

app = Flask(__name__)

# Route to display the welcome message
@app.route("/welcome")
def welcome():
    return "<h2>Welcome to ABC Corporation</h2>"

# Route to display the details
@app.route("/")
def company_details():
    return "Company Name: ABC Corporation <br> Location: India <br> Contact Detail: 999-999-9999"             

if __name__ == "__main__":
    app.run(host="0.0.0.0")
```

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

```python
"""
Answer:

1. The 'url_for()' function is used for building a URL to the specific function dynamically.
2. Using the 'url_for()' function the user can avoid hardcoding URLs in the templates and codes.
3. Following is the python code to demonstrate working of 'url_for()':
"""

from flask import Flask, url_for

app = Flask(__name__)

@app.route("/home")
def home():
    return "Welcome to the home page"

@app.route("/about")
def about():
    return "This is About Page"

@app.route("/")
def index():
    home_url = url_for("home")
    about_url = url_for("about")
    return f"""
    <h2>Navigation</h2>
    <ul>
        <li><a href="{home_url}">Home</a></li>
        <li><a href="{about_url}">About</a></li>
    </ul>
    """

if __name__ == "__main__":
    app.run(host="0.0.0.0")
```
